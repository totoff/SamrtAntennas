```python
import numpy as np
import matplotlib.pyplot as plt
mu0 = 4*np.pi*1E-7
epsilon0 = 8.85E-12
```

```python
#  Plot the ratio of alpha over beta where the loss tangent varies from 0 to 1000.

loss_tangent = np.arange(0.001,1000,0.001)
y=np.sqrt((np.sqrt(1+np.square(loss_tangent))-1)/(np.sqrt(1+np.square(loss_tangent))+1))

plt.semilogx(loss_tangent, y, 'k',.01,1,'k--',100,1,'k--')
plt.xlabel(r'$\sigma/\omega\epsilon$')
plt.ylabel(r'$\alpha / \beta$')
plt.title(r'ratio of attenuation constant ($\alpha$) to phase constant ($\beta$)')
plt.vlines(x=0.01,ymin=-0.05,ymax=1.05,colors='k', linestyles='--')
plt.text(3E-3,.4,'Good insulator',fontsize=10,rotation=90)
plt.vlines(x=100,ymin=-0.05,ymax=1.05,colors='k', linestyles='--')
plt.text(300,.4,'Good conductor',fontsize=10,rotation=90)
plt.ylim(-0.05,1.05)
plt.show()
```

```python
# Plot of molecular resonance of H20 for frequencies above 10GHz

import math
f= np.arange(10,100,0.25)
P=1013
T=300
pv=7.5
gam1=(P/1013)*((300/T)**.626)*(1+.018*pv*T/P)
G=1/(np.square((494.4-np.square(f)))+4*np.square(f)*gam1**2)
k=2*pv*(300/T)**1.5*gam1*np.square(f)*((300/T)*math.exp(-644/T)*G+1.2E-6)
plt.semilogy(f,k/2,'k')
plt.xlabel('Frequency (GHz)')
plt.xticks(np.arange(10,110,10))
plt.ylabel('Specific Attenuation (dB/km)')
plt.title('molecular resonance of H20 for frequencies above 10GHz')
plt.ylim(1E-4,10)
```

```python
# General envelope of standing wave for planewaves 
# normally incident on a planar boundary .  Medium 1 is assumed to be free space
# normalized impedance ...impedance divided => sqrt(mu0/epsilon0)
imp=1/2
G=(imp-1)/(imp+1)
angg=np.arctan2(np.imag(G),np.real(G))
bz=np.arange(-4*np.pi,0,.001)
env=np.sqrt(1+abs(G)**2+2*abs(G)*np.cos(2*bz+angg))
plt.plot(bz,env,'k')
plt.xlabel(r'$\beta_z$')
plt.ylabel(r'$|E_1|$')
plt.ylim(0,1.5)
plt.grid(True)
plt.show()
```

```python
# Calculate the fresnel coefficients for Parallel polarization
# both-media are lossless and epsilon1=epsilon0

theta= np.arange(0,np.pi/2,0.01)
epsr=2
thetat=np.arcsin(np.sqrt(1/epsr)*np.sin(theta))

den=np.cos(thetat)+np.sqrt(epsr)*np.cos(theta)
g2=(np.cos(thetat)-np.sqrt(epsr)*np.cos(theta))/den
T2=(1+g2)*np.cos(theta)/np.cos(theta)

epsr=8
thetat=np.arcsin(np.sqrt(1/epsr)*np.sin(theta))
den=np.cos(thetat)+np.sqrt(epsr)*np.cos(theta)
g3=(np.cos(thetat)-np.sqrt(epsr)*np.cos(theta))/den
T3=(1+g3)*np.cos(theta)/np.cos(theta)

epsr=32
thetat=np.arcsin(np.sqrt(1/epsr)*np.sin(theta))
den=np.cos(thetat)+np.sqrt(epsr)*np.cos(theta)
g4=(np.cos(thetat)-np.sqrt(epsr)*np.cos(theta))/den
T4=(1+g4)*np.cos(theta)/np.cos(theta)

f, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 5))
ax1.plot(np.degrees(theta),abs(g2),'k',np.degrees(theta),abs(g3),'k',np.degrees(theta),abs(g4),'k')
ax1.text(20,.22,r'$\epsilon_{r_2} = 2$')
ax1.text(30,.49,r'$\epsilon_{r_2} = 8$')
ax1.text(40,.69,r'$\epsilon_{r_2} = 32$')
ax1.set_xticks(np.arange(0,100,10))
ax1.set_xlim(0,90)
ax1.set_yticks(np.arange(0,1.2,0.2))
ax1.set_ylim(0,1)
ax1.set_xlabel(r'$\theta$')
ax1.set_ylabel(r'$|R_{||}|$')
ax1.set_title('Reflection Coefficient for Parallel Polarization')


ax2.plot(np.degrees(theta),abs(T2),'k',np.degrees(theta),abs(T3),'k',np.degrees(theta),abs(T4),'k')
ax2.text(35,1.05,r'$\epsilon_{r_2} = 2$')
ax2.text(42,.75,r'$\epsilon_{r_2} = 8$')
ax2.text(50,.55,r'$\epsilon_{r_2} = 32$')
ax2.set_xticks(np.arange(0,100,10))
ax2.set_xlim(0,90)
ax2.set_yticks(np.arange(0,2.5,0.5))
ax2.set_ylim(0,2)
ax2.set_xlabel(r'$\theta$')
ax2.set_ylabel(r'$|T_{||}|$')
ax2.set_title('Transmission Coefficient  for Parallel Polarization')
plt.tight_layout()
```

```python
# Calculate the fresnel coefficients for Perpindicular polarization
# both-media are lossless and epsilon1=epsilon0

theta= np.arange(0,np.pi/2,0.01)
epsr=2
thetat=np.arcsin(np.sqrt(1/epsr)*np.sin(theta))
den=np.cos(theta)+np.sqrt(epsr)*np.cos(thetat)
g2=(np.cos(theta)-np.sqrt(epsr)*np.cos(thetat))/den
T2=(1+g2)

epsr=8;
thetat=np.arcsin(np.sqrt(1/epsr)*np.sin(theta))
den=np.cos(theta)+np.sqrt(epsr)*np.cos(thetat)
g3=(np.cos(theta)-np.sqrt(epsr)*np.cos(thetat))/den
T3=(1+g3)

epsr=32
thetat=np.arcsin(np.sqrt(1/epsr)*np.sin(theta))
den=np.cos(theta)+np.sqrt(epsr)*np.cos(thetat)
g4=(np.cos(theta)-np.sqrt(epsr)*np.cos(thetat))/den
T4=(1+g4)

f, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 5))
ax1.plot(np.degrees(theta),abs(g2),'k',np.degrees(theta),abs(g3),'k',np.degrees(theta),abs(g4),'k')
ax1.text(55,.26,r'$\epsilon_{r_2} = 2$')
ax1.text(40,.45,r'$\epsilon_{r_2} = 8$')
ax1.text(20,.65,r'$\epsilon_{r_2} = 32$')
ax1.set_xticks(np.arange(0,100,10))
ax1.set_xlim(0,90)
ax1.set_yticks(np.arange(0,1.2,0.2))
ax1.set_ylim(0,1)
ax1.set_xlabel(r'$\theta$')
ax1.set_ylabel(r'$|R_\perp|$')
ax1.set_title('Reflection Coefficient for Perpendicular Polarization')


ax2.plot(np.degrees(theta),abs(T2),'k',np.degrees(theta),abs(T3),'k',np.degrees(theta),abs(T4),'k')
ax2.text(55,.76,r'$\epsilon_{r_2} = 2$')
ax2.text(40,.55,r'$\epsilon_{r_2} = 8$')
ax2.text(20,.38,r'$\epsilon_{r_2} = 32$')
ax2.set_xticks(np.arange(0,100,10))
ax2.set_xlim(0,90)
ax2.set_yticks(np.arange(0,1.2,0.2))
ax2.set_ylim(0,1)
ax2.set_xlabel(r'$\theta$')
ax2.set_ylabel(r'$|T_\perp|$')
ax2.set_title('Transmission Coefficient  for Perpendicular Polarization')
plt.tight_layout()
```

```python
# Knife edge diffraction for a half-plane which is at the distance hc from the line of sight.  
# -hc is for the knife edge below the lign of sight, hc is for above
# use trapezoidal integration
# Hc=hc*sqrt(2*d/(lambda*d1*d2

hcc=-5
inc=.01

height= np.array([])
y = np.array([])

for k in np.arange(1,1001):
    Hc=hcc+k/100
    height= np.append(height,Hc)
    u = np.arange(Hc,90,inc)
    next_y = np.abs(inc*np.trapz(np.exp(-1j*np.pi*(np.square(u))/2)))/np.sqrt(2)
    y = np.append(y, next_y)
    

plt.plot(height,y,'k')
plt.xlabel(r'$H_c$')
plt.xticks(np.arange(-5,6,1))
plt.xlim(-5,5)
plt.ylabel(r'$F_d$')
plt.ylim(0.0,1.5)
plt.yticks(np.arange(0.0,1.75,0.25))
plt.vlines(x=0,ymin=0.0,ymax=1.5,colors='k', linestyles='--')
plt.text(-4,.4,'Lit Region',fontsize=10)
plt.text(2,.4,'Shadow Region',fontsize=10)
plt.show()
```

```python
#Pb 2.2
mu=6*mu0
epsilon=epsilon0
alpha10MHz = 1
f=10E6
omega = (2*np.pi*f)
x = ((alpha10MHz/omega)**2)*2/(mu*epsilon)+1
beta10MHz = omega*np.sqrt((mu*epsilon)/2*(x+1))
print(f"beta 10 MHz: {beta10MHz}")
loss_tangent = np.sqrt(x**2-1)
print(f"loss tangent 10 MHz: {loss_tangent}")
sigma10MHz = loss_tangent*omega*epsilon
print(f"sigma 10 MHz: {sigma10MHz}")
impedance=np.sqrt((mu/epsilon)/1-1.j*loss_tangent)
print(f"impedance 10 MHz: {impedance}")
```
